### Imports and other housekeeping

```python
import os
import numpy as np
import pickle
import torch
from matplotlib import pyplot as plt
```

```python
tasks = ['walk', 'back_flip']
strategy = ['exhaustive', 'adaptive', 'cross_entropy']
model_type = ['base_heavy', 'base_rear_heavy', 'base_rear_heavier']
err = {}
for t in tasks:
    err[t] = {}
    for mt in model_type: 
        err[t][mt] = {}
        for st in strategy:
            path = f'{t}/{st}_{mt}/errors/iteration.pkl'
            try:
                with open(path, 'rb') as f:
                    data  = pickle.load(f)
                err[t][mt][st] = [list(range(len(data))), [j if j != 999999 else np.nan for j in [i.cpu() if torch.is_tensor(i) else i for i in data]]]
            except:
                pass
```

### Walk with a heavy base

```python
for k, v in err['walk']['base_heavy'].items():
    plt.plot(v[0], v[1], label=k)
plt.legend()
plt.ylabel('error per iteration')
```

### Walk with a heavy base and heavy rear (thighs and hip)

```python
for k, v in err['walk']['base_rear_heavy'].items():
    plt.plot(v[0], v[1], label=k)
plt.legend()
plt.ylabel('error per iteration')
```

### Walk with a heavier base and heavier rear (thighs and hip)

```python
for k, v in err['walk']['base_rear_heavier'].items():
    plt.plot(v[0], v[1], label=k)
plt.legend()
plt.ylabel('error per iteration')
```
